```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/company_sales_data.csv')

profitList = df['total_profit'].tolist()

monthList = df ['month_number'].tolist()

plt.plot(monthList,profitList,label='profit of company',color='b',marker='^')

plt.xlabel('Month number')
plt.ylabel('Profit of company')
# plt.yticks(profitList)

plt.xticks(monthList)
plt.title('Company profit per month',color='r')
plt.yticks([100000,200000,300000, 400000,500000])
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/company_sales_data.csv')

profitlist = df ['total_profit'].tolist()
#labels = ['low', 'average', 'Good]

profit_range=[150000,175000,200000,225000,300000,350000]
plt.hist(profitlist,profit_range, rwidth=0.60, label='profit data')

plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')

plt.show()
```
